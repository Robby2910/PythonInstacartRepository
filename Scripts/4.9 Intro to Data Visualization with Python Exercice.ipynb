# 04. Creating Histograms & Scatterplots

```python
# The code below will create an histogram
df_ords_prods_group_aggr['prices'].plot.hist(bins = 25)
```

```python
# The code below will use a scatterplot to investigate the data and see if there are any outliers
sns.scatterplot(x = 'prices', y = 'prices',data = df_ords_prods_group_aggr)
```

```python
# The code below will check for values above 100 in our dataframe
df_ords_prods_group_aggr.loc[df_ords_prods_group_aggr['prices'] > 100].drop(['Unnamed: 0_x', 'Unnamed: 0.1', 'Unnamed: 0_y', 'mean_order'], axis = 1)
```

```python
# The code below will turn all value above 100 into missing value
df_ords_prods_group_aggr.loc[df_ords_prods_group_aggr['prices'] >100, 'prices'] = np.nan
```

```python
# The code below will check for the max value again to see if the cleaning and correction worked
df_ords_prods_group_aggr['prices'].max()
```

```python
# And now the code below will run the histogram again to see if the changes have been made
df_ords_prods_group_aggr['prices'].plot.hist(bins = 70)
```
