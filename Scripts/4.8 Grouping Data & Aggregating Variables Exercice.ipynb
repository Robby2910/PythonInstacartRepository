```python
df_subset_merge.groupby('product_name')
```

```python
# the code below will group and aggregate data by combining the groupby() and agg() function
```

```python
df_subset_merge.groupby('department_id').agg({'order_number': ['mean']})
```

```python
# The code below will give the same result as the code above, but without the use of the agg() function
df_subset_merge.groupby('department_id')['order_number'].mean()
```

```python
# The code below will perform a multiple aggregation
df_subset_merge.groupby('department_id').agg({'order_number': ['mean', 'min', 'max']})
```

```python
# The code below will help us create a new column with the transform() function
df_subset_merge['max_order'] = df_subset_merge.groupby(['user_id'])['order_number'].transform(np.max)
```

```python
df_subset_merge.head(15).drop(['Unnamed: 0_x', 'Unnamed: 0.1', 'Unnamed: 0_y'], axis = 1)
```

```python
# The codes below will create a flag from the newly created column 'max_order'
```

```python
df_subset_merge.loc[df_subset_merge['max_order'] > 40, 'loyalty_flag'] = 'Loyal customer'
```

```python
df_subset_merge.loc[(df_subset_merge['max_order'] <= 40) & (df_subset_merge['max_order'] > 10), 'loyalty_flag'] = 'Regular customer'
```

```python
df_subset_merge.loc[df_subset_merge['max_order'] <= 10, 'loyalty_flag'] = 'New customer'
```
