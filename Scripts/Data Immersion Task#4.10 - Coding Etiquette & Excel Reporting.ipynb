```python
# 05. The marketing and business strategy units at Instacart want to create more-relevant marketing strategies 
# for different products and are, thus, curious about customer profiling in their database. Create a profiling 
# variable based on age, income, certain goods in the “department_id” column, and number of dependants. 
# You might also use the “orders_day_of_the_week” and “order_hour_of_day” columns if you can think of a way 
# they would impact customer profiles. (Hint: As an example, try thinking of what characteristics would lead you to 
# the profile “Single adult” or “Young parent.”)
```

```python
# The code below will use the .loc() function to create a customer profile 'single adult', 'young parent', or 'older parent'
# based of the characteristics 'Age' and '#_of_dependants'.        
```

```python
df_report_excluded.loc[(df_report_excluded['#_of_dependants'] <= 0) & (df_report_excluded['Age'] <= 30), 'cust_profiling'] = 'single adult'
```

```python
df_report_excluded.loc[(df_report_excluded['#_of_dependants'] >= 1) & (df_report_excluded['Age'] <= 30), 'cust_profiling'] = 'young parent'
```

```python
df_report_excluded.loc[(df_report_excluded['#_of_dependants'] >= 1) & (df_report_excluded['Age'] >= 30), 'cust_profiling'] = 'older parent'
```
