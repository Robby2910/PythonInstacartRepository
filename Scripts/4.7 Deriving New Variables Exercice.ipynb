```python
# the code below will make a subset of the above dataframe for the first 1 millions row
df_subset = df_ords_prods_merged[:1000000]
```

```python
df_subset.drop(['Unnamed: 0_x', 'Unnamed: 0.1', 'Unnamed: 0_y'], axis = 1)
```

```python
df_subset.shape
```

```python
# the code below will define a function
def price_label(row):

  if row['prices'] <= 5:
    return 'Low-range product'
  elif (row['prices'] > 5) and (row['prices'] <= 15):
    return 'Mid-range product'
  elif row['prices'] > 15:
    return 'High range'
  else: return 'Not enough data'
```

```python
# The code below will apply the function
df_subset['price_range'] = df_subset.apply(price_label, axis = 1)
```

```python
df_subset['price_range'].value_counts(dropna = False)
```

```python
# The code below will check what is tha max value for 'price 'column
df_subset['prices'].max()
```

```python
# The codes below will use the 'loc()' to come as the same result as the defined function above, but will be faster
df_subset.loc[df_subset['prices'] > 15, 'price_range_loc'] = 'High-range product'
```

```python
df_subset.loc[(df_subset['prices'] <= 15) & (df_subset['prices'] > 5), 'price_range_loc'] = 'Mid-range product'
```

```python
df_subset.loc[df_subset['prices'] <= 5, 'price_range_loc'] = 'Low-range product'
```

```python
df_subset['price_range_loc'].value_counts(dropna = False)
```

```python
# The code below will use the 'loc()' to come up with the same result as above but for the whole
#dataframe instead of just a subset of the dataframe. We leraned that 'loc()' function is easier on memory
```

```python
# the code below will print the frequency for the 'orders_day_of_week'column
df_ords_prods_merged['orders_day_of_week'].value_counts(dropna = False)
```

```python
# The code below will create a for loop that will create a new column indicating how busy the day of the week 
# is from the 'orders_day_of_week' column
result = []

for value in df_ords_prods_merged["orders_day_of_week"]:
  if value == 0:
    result.append("Busiest day")
  elif value == 4:
    result.append("Least busy")
  else:
    result.append("Regularly busy")
```

```python
result
```

```python
# The code below will cresate a new column called 'busiest_day' that will be added to the dataframe 
df_ords_prods_merged['busiest_day'] = result
```

```python
df_ords_prods_merged['busiest_day'].value_counts(dropna = False)
```
